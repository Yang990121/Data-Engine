# Import and inspect api generated csvs

```python

# Specify the file path
file_path_2 = "resale_flat_prices_1990-1999_new.csv"
file_path_3 = "resale_flat_prices_2000-2012_new.csv"
file_path_4 = "resale_flat_prices_2012-2014_new.csv"
file_path_5 = "resale_flat_prices_2015-2016_new.csv"
file_path_6 = "resale_flat_prices_2017-2024_new.csv"

# Read the CSV file into a DataFrame
df_2 = pd.read_csv(file_path_2)
df_3 = pd.read_csv(file_path_3)
df_4 = pd.read_csv(file_path_4)
df_5 = pd.read_csv(file_path_5)
df_6 = pd.read_csv(file_path_6)

# Display the first few rows of the DataFrame to verify import
print(len(df_2))  # Print the length of the df_2
df_2.head()
```

# Combining the dfs

```python
# Concatenate them along rows
combined_df = pd.concat([df_2, df_3, df_4, df_5, df_6], ignore_index=True)

print(len(combined_df))  # Print the length of the combined DataFrame

combined_df.head()  # Display the first few rows of the combined DataFrame
```

# Processing of columns

```python
# Remove the 'id' column since id is duplicated for different dfs
if '_id' in combined_df.columns:
    combined_df.drop(columns=['_id'], inplace=True)
    
combined_df.head()  # Display the first few rows of the combined DataFrame
```

```python
# Find the number of null values in the 'remaining_lease' column
num_null_remaining_lease = combined_df['remaining_lease'].isnull().sum()

print("Number of null values in 'remaining_lease' column:", num_null_remaining_lease)
```

```python
# Remove the 'remaining_lease' column as the proportion of null values in the df is too large
if 'remaining_lease' in combined_df.columns:
    combined_df.drop(columns=['remaining_lease'], inplace=True)
    
combined_df.head()  # Display the first few rows of the combined DataFrame
```
