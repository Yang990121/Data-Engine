```python
import pandas as pd
```

# Append feature dictionary to combined dataframe

```python
csv_file_path = 'combined_dataframe.csv'
df = pd.read_csv(csv_file_path)
print(len(df))
df.head()
```

```python
dict_file_path = 'feature_dict.csv'
dict = pd.read_csv(dict_file_path)

print(len(dict))
dict.head()
```

```python
# Create a new column by concatenating 'block' and 'street_name'
df['address'] = df['block'] + ', ' + df['street_name']

# Find the number of distinct addresses in the DataFrame
num_distinct_addresses = df['address'].nunique()

# Print the number of distinct addresses
print("Number of distinct addresses:", num_distinct_addresses)

df.head()
```

```python
# Find unique addresses in df['address']
unique_addresses_combined = set(df['address'])

# Find unique addresses in df['address']
unique_addresses_df = set(dict['address'])

# Find unique addresses in combined_df['address'] but not in df['address']
unique_addresses_only_combined = unique_addresses_combined - unique_addresses_df

# Print unique addresses only in combined_df['address'] but not in df['address']
for address in unique_addresses_only_combined:
    print(address)
```

```python
# Perform an inner join between df and feature_dict on the 'address' column
merged_df = pd.merge(df, dict, on='address', how='inner')

# Display the merged DataFrame
merged_df.head()
```

```python
# Check for null values in the merged DataFrame
null_values = merged_df.isnull().sum()

# Display columns with null values
print("Columns with null values:")
print(null_values[null_values > 0])
```
