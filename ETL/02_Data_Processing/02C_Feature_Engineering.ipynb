```python
import pandas as pd
```

# Append feature dictionary to combined dataframe

```python
csv_file_path = 'combined_dataframe.csv'
df = pd.read_csv(csv_file_path)
print(len(df))
df.head()
```

```python
dict_file_path = 'feature_dict.csv'
dict = pd.read_csv(dict_file_path)

print(len(dict))
dict.head()
```

```python
# Create a new column by concatenating 'block' and 'street_name'
df['address'] = df['block'] + ', ' + df['street_name']

# Find the number of distinct addresses in the DataFrame
num_distinct_addresses = df['address'].nunique()

# Print the number of distinct addresses
print("Number of distinct addresses:", num_distinct_addresses)

df.head()
```

```python
# Find unique addresses in df['address']
unique_addresses_combined = set(df['address'])

# Find unique addresses in df['address']
unique_addresses_df = set(dict['address'])

# Find unique addresses in combined_df['address'] but not in df['address']
unique_addresses_only_combined = unique_addresses_combined - unique_addresses_df

# Print unique addresses only in combined_df['address'] but not in df['address']
for address in unique_addresses_only_combined:
    print(address)
```

```python
# Perform an inner join between df and feature_dict on the 'address' column
merged_df = pd.merge(df, dict, on='address', how='inner')

# Display the merged DataFrame
merged_df.head()
```

```python
# Check for null values in the merged DataFrame
null_values = merged_df.isnull().sum()

# Display columns with null values
print("Columns with null values:")
print(null_values[null_values > 0])
```

# Feature engineering from merged df

```python
merged_df.head()
```

## Process story_range

```python
# Print out the unique values in the 'storey_range' column
unique_storey_ranges = merged_df['storey_range'].unique()

print("Unique values in the 'storey_range' column:")
print(unique_storey_ranges)
```

```python
# Clean the storey_range data and calculate average range
avg_storey_ranges = {}

for range_str in unique_storey_ranges:
    start, end = map(int, range_str.split(' TO '))
    avg_storey = (start + end) / 2
    avg_storey_ranges[range_str] = avg_storey
    
print(avg_storey_ranges)
```

```python
# Add a new column 'avg_storey_range' using the dictionary mapping
merged_df['avg_storey_range'] = merged_df['storey_range'].map(avg_storey_ranges)
```

```python
merged_df.head()
```

## Process month feature

```python
# Split 'month' column into 'year' and 'quarter'
merged_df[['year', 'quarter']] = merged_df['month'].str.split('-', expand=True)

# Define a function to map month to quarter
def map_to_quarter(month):
    month_int = int(month)
    if 1 <= month_int <= 3:
        return '1'
    elif 4 <= month_int <= 6:
        return '2'
    elif 7 <= month_int <= 9:
        return '3'
    else:
        return '4'

# Apply the function to map month to quarter
merged_df['quarter'] = merged_df['quarter'].apply(map_to_quarter)

merged_df.head()
```

## Combining features to add interaction

```python
# Combining flat_type and flat_model

# Combine 'flat_type' and 'flat_model' columns into a new column 'flat_type_model'
merged_df['flat_type_model'] = merged_df['flat_type'] + ' ' + merged_df['flat_model']

# Print the modified DataFrame
merged_df.head()
```
