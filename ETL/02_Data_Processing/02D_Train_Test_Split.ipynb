```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final.csv')
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
df1 = df.copy()
```

### Preparation

```python
df1.columns
```

```python
# sort by month
df1 = df1.sort_values('month')
df1.head()
```

take the first 70% of the data as training data

this is done because there should not be any 'interpolation' between the training and test data, i.e. the training data should not have any information about the future, due to the underlying time series nature of the data.

```python
# take the first 70% of the data as training data
# take the next 15% as validation data
# take the last 15% as test data

train = df1.head(int(len(df1)*0.7))
test_tmp = df1.drop(train.index)
validation = test_tmp.head(int(len(test_tmp)*0.5))
test = test_tmp.drop(validation.index)
```

```python
# print the proportion of each dataset to the total
print('train:', len(train)/len(df1))
print('validation:', len(validation)/len(df1))
print('test:', len(test)/len(df1))
```

```python
# save the data to csv
train.to_csv('train.csv', index=False)
validation.to_csv('validation.csv', index=False)
test.to_csv('test.csv', index=False)
```
